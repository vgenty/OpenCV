```python
#Canny gets worse as increase blur...
n=10
for i in xrange(0,n):
    #ax = plt.subplot(n,1,i+1)
    fig = plt.subplots(figsize=(10,6))
    img = cv2.blur(image,(i+1,i+1))
    ax = plt.subplot(1,2,1)
    ax.imshow(img,cmap = cm.Greys,interpolation='none')
    ax = plt.subplot(1,2,2)
    ax.imshow(cv2.Canny(img,10,100,3),cmap = cm.Greys,interpolation='none')
    plt.show()
```
